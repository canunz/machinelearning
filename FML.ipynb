Un gráfico relacionado a los cuartiles y describe varias características importantes al mismo tiempo, tales como la dispersión y simetría es el diagrama de caja. Para su realización se representan los tres cuartiles y los valores mínimo y máximo de los datos, sobre un rectángulo, alineado horizontal o verticalmente.

Este gráfico indica.

image.png

Mediana: el valor medio de los datos. marcado como Q2, representa el percentil 50. Primer cuartil: el valor medio entre el "mínimo no atípico" y la mediana. marcado como Q1, representa el percentil 25. Tercer cuartil: el valor medio entre “máximo no atípico” y la mediana. marcado como Q3, representa el percentil 75. “Máximo no atípico”: calculado por(Q3 + 1.5 * IQR). Todos los valores por encima de este se consideran valores atípicos. “Mínimo no atípico”: calculado por(Q1 - 1,5 * IQR). Todos los valores por debajo de este se consideran valores atípicos.


Un gráfico de cajas (también llamado boxplot) es una herramienta visual que muestra la distribución de un conjunto de datos. Te ayuda a entender rápidamente la variabilidad, la simetría, y la presencia de posibles valores atípicos (outliers). Aquí te explico cómo leer un gráfico de cajas.

Partes de un gráfico de cajas:
Caja (box):

Representa el rango intercuartílico (IQR), que abarca el 50% central de los datos.

La caja está delimitada por el primer cuartil (Q1) (el límite inferior) y el tercer cuartil (Q3) (el límite superior).

Q1 (25% de los datos) es el valor en el que el 25% de los datos son menores.

Q3 (75% de los datos) es el valor en el que el 75% de los datos son menores.

Mediana (línea dentro de la caja):

Es el valor central de los datos, donde el 50% de los datos están por debajo y el 50% están por encima.

Se muestra como una línea horizontal dentro de la caja.

Bigotes (whiskers):

Los bigotes se extienden desde los cuartiles hasta el valor máximo o mínimo que aún se considera "normal".

Generalmente, los bigotes se extienden hasta 1.5 veces el rango intercuartílico (1.5 * IQR) por fuera de los cuartiles.

El límite superior es Q3 + 1.5 * IQR, y el límite inferior es Q1 - 1.5 * IQR.

Valores atípicos (outliers):

Los puntos fuera de los bigotes (es decir, aquellos que están más allá de 1.5 * IQR) son valores atípicos. Estos puntos se marcan generalmente con círculos o asteriscos.

Los valores atípicos pueden indicar que algunos datos se alejan significativamente de la tendencia central y pueden necesitar un análisis más detallado.

Ejemplo de cómo leer el gráfico:
Supongamos que tenemos un boxplot que muestra la distribución de salarios:

Caja: La caja muestra el 50% de los salarios. Si está ubicada en un valor alto, indica que la mayoría de los salarios están en ese rango superior.

Mediana: Si la línea dentro de la caja está más cerca del borde inferior, significa que la distribución es sesgada hacia los valores más bajos.

Bigotes: Si los bigotes son largos, eso indica que los datos se extienden mucho más allá del rango intercuartílico, lo que podría indicar una alta variabilidad.

Outliers: Los puntos fuera de los bigotes indican personas con salarios que están muy por fuera de la norma, posiblemente por ser extremadamente bajos o altos en comparación con el resto de los datos.
